```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = [
    "Hello! How are you doing today?",
    "I hope you’re having a great day.",
    "It’s been a while since we last talked.",
    "Do you have any plans for the weekend?",
    "I just wanted to check in and say hi!",
    "Let me know if you need any help with your project.",
    "What’s your favorite book or movie?",
    "I’m learning new things every day, and it feels great!"
]
```

```python
### Define vocab size
voc_size=1000
```

```python
## one hot representation
one_hot_repr=[one_hot(words,voc_size)for words in sent]
one_hot_repr
```

```python
## wors Embedding Representaion

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
## try to make each sentence of 12 word
sent_length=12
embedde_doc=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
embedde_doc
```

```python
## freature repensentation 
dim=10
model=Sequential()
model.add(Embedding(input_dim=voc_size, output_dim=dim, input_length=sent_length))
model.compile('adam','mse')

```

```python
model.predict(embedde_doc)
```
